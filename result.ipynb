# ***Config***

```python
drive.mount("/content/drive")
warnings.filterwarnings('ignore')
samples_df = pd.read_csv("SpeechCARE_Acoustic_Explainability_Framework/data/samples_df.csv")
```

```python
model_id = "large-v3"
compute_type = "int8"
CHECKPOINTS_FILE = "checkpoints/model_best.pt"
```

#***Samples***

##***lxbg: ADRD***

```python
AUDIO_PATH = samples_df.loc[0,"path"].split("/")[-1]
DEMOGRAPHIC_INFO = float(samples_df.loc[0,"age"])
FIG_SAVE_PATH = f"SpeechCARE_Acoustic_Explainability_Framework/figs/{samples_df.loc[0,'uid']}_spectrogram.png"
WORD_SEGMENTS_PATH = f"SpeechCARE_Acoustic_Explainability_Framework/data/{samples_df.loc[0,'uid']}_WhisperX_{model_id}_{compute_type}_charLevel.json"
```

```python
!python SpeechCARE_Acoustic_Explainability_Framework/test/test.py --model_checkpoint $CHECKPOINTS_FILE \
                                                                  --audio_path $AUDIO_PATH \
                                                                  --demography_info $DEMOGRAPHIC_INFO \
                                                                  --fig_save_path $FIG_SAVE_PATH \
                                                                  --word_segments $WORD_SEGMENTS_PATH \
                                                                  --min_pause_duration 0.15
```
